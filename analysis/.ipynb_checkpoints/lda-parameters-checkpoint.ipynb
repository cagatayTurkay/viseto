# Topic Modelling for TimeSets

Look at the data first.

```python
import json

import pandas as pd
pd.set_option('display.max_colwidth', 1000)

from gensim.models.ldamodel import LdaModel
from gensim import corpora

from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string
```

```python
def extract_text_fields(post):
    # Which fields? name, message, story, comments -> message
    return { 
        'name': post.get('name'),
        'message': post.get('message'),
        'story': post.get('story'),
        'type': post.get('type'),
        'comments': '. '.join([c.get('message') for c in post['comments']['data']]) if 'comments' in post else None
    }

def load_data(filename):
    with open(filename) as f:
        posts = json.load(f) # list of posts
        return pd.DataFrame(extract_text_fields(post) for post in posts)

df = load_data('../data/detailedFbData.json')
df.head(20)
```

Look like `story` is not meaningful, `message` and `comments` can be combined as a text field for topic modelling, and `name` can be used as the display title in TimeSets.

```python
def build_docs(posts):
    "Return a list of documents."
    return [(p['message'] or '') for p in posts]
```

```python
docs = build_docs(df.to_dict('records'))
docs[1]
```

```python
def clean(doc, stop, exclude, lemma):
    stop_free = ' '.join([w for w in doc.lower().split() if w not in stop])
    punc_free = ''.join(c for c in stop_free if c not in exclude)
    normalized = ' '.join(lemma.lemmatize(w) for w in punc_free.split())
    return normalized

def preprocess(docs):
    "Return a bag-of-word representation of the preprocessed documents."
    stop = set(stopwords.words('english'))
    exclude = set(string.punctuation) 
    lemma = WordNetLemmatizer()
    return [clean(doc, stop, exclude, lemma).split() for doc in docs]
```

```python
clean_docs = preprocess(docs)
clean_docs[0]
```

```python
def build_lda(docs, num_topics=10, passes=10):
    "Return an LDA model from the given list of documents."
    dictionary = corpora.Dictionary(docs)
    doc_term_matrix = [dictionary.doc2bow(doc) for doc in docs]    
    return LdaModel(doc_term_matrix, num_topics=num_topics, id2word=dictionary, passes=passes, random_state=0)

def get_topics(lda):
    return [[(lda.id2word[t], float(format(p, '.2f'))) for t, p in lda.get_topic_terms(i)] for i in range(lda.num_topics)]
```

```python
lda = build_lda(clean_docs, num_topics=20)
```

```python
get_topics(lda)
```
